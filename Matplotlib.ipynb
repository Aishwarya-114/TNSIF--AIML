# Plot graph

Showing continuous data over time.

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [120, 150, 180, 170, 210]
plt.plot(months, sales, marker='o')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(["Sales"])
plt.show()
```

# Histogram

Shows frequency distribution of numerical data.

```python
ages = [18000,19000,20000,20000,21000,22000,22000,23000,24000,25000,25000,25000]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

# Scatter Plot

To find relationships between variables.

```python
height = [150, 155, 160, 165, 170, 175]
weight = [45, 50, 55, 60, 68, 75]

plt.scatter(height, weight)
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

# Pie Chart

Showing percentages.

```python
labels = ['Vivo', 'Apple', 'Redmi', 'Xiaomi']
values = [45, 25, 15, 15]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Mobile Brand Usage")
plt.show()
```

# multi[le plots in one figure

# **Task**

Create a dataset containing marks of
10 students.

students = ['A','B','C','D','E','F','G','H','I','J']

marks = [78,85,92,67,88,95,72,81,90,76]

1.Display student marks graphically.

```python
import matplotlib.pyplot as plt

# Dataset
students = ['A','B','C','D','E','F','G','H','I','J']
marks = [78,85,92,67,88,95,72,81,90,76]

plt.figure(figsize=(8,4))
plt.bar(students, marks)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

2.Find the highest scorer.

```python
highest_marks = max(marks)
highest_student = students[marks.index(highest_marks)]

print("Highest Scorer:", highest_student)
print("Highest Marks:", highest_marks)
```

3.Display marks in descending order and plot again.

```python
sorted_data = sorted(zip(marks, students), reverse=True)

sorted_marks = [m for m, s in sorted_data]
sorted_students = [s for m, s in sorted_data]
print("\nMarks in Descending Order:")
for s, m in zip(sorted_students, sorted_marks):
    print(s, ":", m)
```

```python
plt.figure(figsize=(8,4))
plt.bar(sorted_students, sorted_marks)
plt.title("Student Marks (Descending Order)")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

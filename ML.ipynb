# **Artificial Intelligence (AI)**

Artificial Intelligence (AI) is a branch of computer science that aims to create systems capable of performing tasks that normally require human intelligence.

# Examples of AI
Siri and Google Assistant answering questions.

ChatGPT generating human-like responses.

Self-driving cars making driving decisions.

Chess-playing programs competing against humans.

# **Machine Learning (ML)**

Machine Learning is a subset of Artificial Intelligence.

Instead of programming every rule manually, machine learning enables computers to learn patterns from data and improve their performance automatically.

# Examples of Machine Learning
Email spam detection.

Netflix movie recommendations.

Credit card fraud detection.

Weather prediction.

Disease prediction.

# Types of Machine Learning

1.Supervised Learning

2.Unsupervised Learning

3.Semi-Supervised Learning

4.Reinforcement Learning

# **Deep Learning (DL)**

Deep Learning is a subset of Machine Learning that uses Artificial Neural Networks inspired by the structure of the human brain.

# Examples of Deep Learning
Face recognition systems.

Speech recognition in Alexa and Siri.

Self-driving cars identifying roads and traffic signs.

Google Translate.

ChatGPT and Large Language Models.

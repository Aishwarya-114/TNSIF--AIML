# **Data Cleaning & Preprocessing**

Garbage Data → Garbage Output Clean Data → Better Models

##  What is Data Cleaning?

Data Cleaning is the process of:
- Fixing incorrect data
- Handling missing values
- Removing duplicates
- Standardizing formats

##  Why is Data Preprocessing Important?

Machine Learning models:
- Cannot handle missing or messy data properly
- Require numerical and consistent input


##  Common Data Issues

1. Missing Values → NaN
2. Duplicate Records
3. Incorrect Data Types
4. Inconsistent Text (Mumbai vs mumbai)

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Ram", "Shyam", "Ram", "Sita", None,"Shyam","Shyam","Shyam","Ram"],
    "Age": [25, np.nan, 25, 30, 22,13,17,18,19],
    "City": ["Mumbai", "mumbai", "Mumbai", "Delhi", "Delhi", "Delhi", "Delhi", "Delhi", "Delhi"]
}

df = pd.DataFrame(data)
df
```

```python
# First few rows
print(df.head())
print('this is info()====================')

# Data info
print(df.info())
print('this is describe()====================')
# Summary statistics
print(df.describe())
```

```python
# Check missing values
print(df.isnull())

# Count missing values
df.isnull().sum() #count of missing values
```

```python
# Drop rows with missing values
df_drop = df.dropna()
df_drop
```

```python
#  Best Practice

df["Age"] = df["Age"].fillna(df["Age"].mean())  #targeting the column
df["Name"] = df["Name"].fillna(df["Name"].mode()[0])
#we dont have any missing values in city column

df
```

# **Encoding (Categorical → Numerical)**

```python
df_encoded = pd.get_dummies(df, columns=["City"])
df_encoded
```

# **Duplicate Records**

```python
df_no_dup = df.drop_duplicates()
df_no_dup
```

```python
# Convert all city names to lowercase
df["City"] = df["City"].str.upper()
df2=df.dropna()

df2
```

```python
df2.dropna()
df2["Age"] = df2["Age"].astype(int)
print(df2)
df2.drop_duplicates()
```

# 🌧️ Rainfall Dataset Cleaning

## 🎯 Objective:
Clean and preprocess the dataset `rainfall.csv`

---

## 🔹 Dataset Columns:
- date
- rainfall
- temperature
- humidity
- wind_speed
- weather_condition

Upload CSV File from Local System (Google Colab)

```python
import pandas as pd
import numpy as np


import pandas as pd

# Load the uploaded file
file_name = '/content/rainfall.csv.xls'
df = pd.read_csv(file_name)

df.head()
```

### ✅ Operations:
1. Inspect data
2. Convert date column
3. Handle missing values
4. Fix text inconsistencies
5. Remove duplicates
6. Handle outliers
7. Feature engineering
8. Encoding categorical data
9. Save cleaned dataset

```python
df.head()

df.info()

df.isnull().sum()
```

```python
df["date"] = df["date"].astype(str).str.strip()
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df.info()
```

```python
import pandas as pd

# Numerical columns
num_cols = ["rainfall", "temperature", "humidity", "wind_speed"]

# Fill numerical missing values with mean
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Fill categorical missing values with mode
df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0])

# Check missing values
print(df.isnull().sum())
```

```python
df["weather_condition"] = df["weather_condition"].str.lower()

df["weather_condition"].unique()

df
```

```python
df.to_csv("cleaned_rainfall.csv", index=False)
```

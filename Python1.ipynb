```python
# Checking whether a number is prime or not

num = int(input("Enter a number: "))
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    print("It is a prime number.")
else:
    print("It is not a prime number.")
```

```python
#factorial

#M1.Factorial of a fixed number
digit = 5
fact = 1
for i in range(1, digit + 1):
    fact = fact * i
print(fact)

 #OR

 # Factorial using user input

n = int(input("Enter a number: "))
fact = 1
for i in range(1, n + 1):
    fact = fact * i
print("Factorial of", n, "is", fact)
```

```python
#Reverse of a number
n=int(input("enter a number: "))
rev=0
while(n>0):
  r=n%10
  rev=rev*10+r
  n=n//10
print("Reverse of n is: ",rev)
```

```python
#Palindrome
n=int(input("enter a number:"))
pal=n
rev=0
while(n>0):
  r=n%10
  rev=rev*10+r
  n=n//10
if (pal==rev):
  print("palindrome")
else:
  print("not palindrome")
```

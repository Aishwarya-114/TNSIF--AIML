# **Datatypes in Python**

```python
#List[]
stds=["Aishu","sushma","karthi","teena"]

print(stds)
print(type(stds))
print("First Fruit:", stds[0])
```

```python

# Tuple Data Type
numbers = (10, 20, 30, 40)

print(numbers)
print(type(numbers))
print("Second Element:", numbers[1])
```

```python
# Set Data Type

colors = {"Red", "Blue", "Green", "Yellow"}

print(colors)
print(type(colors))
```

```python

# Dictionary Data Type

student = {
    "Name": "Aishwarya",
    "Age": 20,
    "Branch": "CSE"
    }

print(student)
print(type(student))
```

```python
# Type Conversion

num = "100"

print("Before Conversion:", type(num))

num = int(num)

print("After Conversion:", type(num))
```

```python
#string
s='python programming'
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.strip())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith('python'))
print(s.endswith('python'))
print(s)
print(len(s))
```

# **Functions in Python**



```python
def function_name():
    # Function body
    print("Hello World")

function_name()
```

```python
#1.Function Without Parameters and Without Return Value

def display():
    print("Learning Python Functions")

display()
```

```python
#2.Function With Parameters

def greet(name):
    print("Hello", name)

greet("Aishwarya")
greet("Karthik")
```

```python
#3.Function With Return Value

def add(a, b):
    return a + b

result = add(10, 20)
print(result)
```

```python
#4.Function With Parameters and Return Value
def square(num):
    return num * num

result = square(5)
print(result)
```

```python
#lambda function
add = lambda a, b: a + b
print(add(2, 3))
```

```python
names=lambda name:name
print(names('AISHU'))
```

```python
#recursive

def factorial(n):
  if n==1:
    return 1
  return n*factorial(n-1)
print(factorial(5))
```

# **OOP in Python**

```python
#class and object
class student:
  sid=101
  sname='karthik'
obj=student()
print(obj.sid)
print(obj.sname)
```

```python
#Constructor
class marks:
  def __init__(self,maths,english,telugu):
    self.maths=maths
    self.english=english
    self.telugu=telugu
m1=marks(84,96,98)
print("maths marks is: ",m1.maths)
```

```python


class student:
  def __init__(self,name,rollno,marks):
    self.name=name
    self.rollno=rollno
    self.marks=marks
  def display(self):
    print(self.name,self.rollno,self.marks)
s=student("Aishu",101,98)
s.display()
```

```python
class rectangle:
  def __init__(self,l,b):
    self.l=l
    self.b=b
  def area(self):
    return self.l*self.b
  def perimeter(self):
    return 2*(self.l+self.b)
r=rectangle(10,5)
print(r.area())
print(r.perimeter())
```

```python
class BankAccount:
    def __init__(self, acc_no, balance):
        self.acc_no = acc_no
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def display_balance(self):
        print("Balance =", self.balance)

acc = BankAccount(12345, 1000)

acc.deposit(500)
acc.withdraw(200)
acc.display_balance()
```

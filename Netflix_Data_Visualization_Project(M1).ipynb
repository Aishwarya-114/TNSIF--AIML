Task to perform

*   Which type of content is more available on Netflix?
*   Which countries contribute the most content to Netflix?
*   Which directors have the most titles on Netflix?






```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/netflix_titles.csv.zip")
df.info()
df.tail()
df.head()
```

```python
data=df.dropna()
data.head()
```

```python
type_counts = data['type'].value_counts()
plt.figure(figsize=(5,6))
plt.pie(type_counts,labels=type_counts.index,autopct='%1.1f%%')

plt.title('Most watched content on Netflix')
plt.show()
```

```python
country_counts = (data['country'].dropna().str.split(', ').explode().value_counts().head(10))
country_data = country_counts.reset_index()
country_data.columns = ['Country', 'Count']
plt.figure(figsize=(6, 4))
sns.barplot(data=country_data, x='Country', y='Count')
plt.title('Top 10 Countries Contributing Content to Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

```python
director_counts = data['director'].dropna().value_counts().head(5)
director_data = director_counts.reset_index()
director_data.columns = ['Director', 'Count']
plt.figure(figsize=(12,4))
sns.barplot(data=director_data, x='Count', y= 'Director')
plt.title('Top 5 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

# List of Functions and Their Usage

 **1)np.array()**-	Create an array

**4)np.ones((m,n))**-	Create array filled with ones

**6)np.linspace(start, stop, num)**-	Create evenly spaced numbers

**7)np.random.rand(m,n)**-	Random numbers between 0 and 1

**8)np.shape**-	Get dimensions of array

**9)np.reshape()**-	Reshape array

**10)np.sum(), np.mean(), np.std()**-	Aggregation functions

**11)np.min(), np.max(), np.argmin(), np.argmax()**-	Min/Max values & indices

***12)np.dot(), np.matmul()***-	Matrix multiplication

**13)np.transpose()**-	Transpose of matrix

**14)np.sort()**- Sorting array

```python

# Importing NumPy
import numpy as np

# Creating Arrays
arr1 = np.array([1,2,3,4,5,6])
arr2 = np.array([[1, 2, 3], [4, 5, 6],[9,10,11],[11,12,13]])

print("1D Array:", arr1)
print("2D Array:\n", arr2)

# Array with specific values
zeros = np.zeros((3,4))
ones = np.ones((2,3))
identity = np.eye(5)

print("Zeros:\n", zeros)
print("Ones:\n", ones)
print("Identity Matrix:\n", identity)

# Range and linspace
arange_arr = np.arange(1, 45, 2)
linspace_arr = np.linspace(1, 2, 4)

print("Arange:", arange_arr)
print("Linspace:", linspace_arr)

# Random numbers
rand_arr = np.random.rand(2,3)
print("Random Array:\n", rand_arr)
otp =( (rand_arr*1000000)).astype(int)
print("Random Array:\n", otp)
```

```python
# Example 1: Reshape and Transpose
matrix = np.arange(1, 13).reshape(4,3)
print("Matrix:\n", matrix)
print("Transpose:\n", matrix.T)

# Example 2: Mathematical operations
arr = np.array([10, 20, 30, 40])
print("Original Array:", arr)
print("Array + 5:", arr - 5)
print("Array squared:", arr ** 2)

# Example 3: Aggregate Functions
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))

# Example 4: Matrix multiplication
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("Matrix Multiplication:\n", np.dot(A,B))

# Example 5: Sorting
unsorted = np.array([3,1,4,1,5,9])
print("Sorted Array:", np.sort(unsorted))
```

```python
import numpy as np

# Create an array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Number of dimensions
print("Number of dimensions:", arr.ndim)

# Shape of the array
print("Shape:", arr.shape)

# Size of the array
print("Size:", arr.size)

# Data type of the array
print("Data type:", arr.dtype)

# Multiply every element by 2
result = arr * 2
print("Array after multiplying by 2:")
print(result)
```

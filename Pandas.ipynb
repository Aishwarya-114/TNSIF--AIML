Pandas is powerful library for data manipulation and analysis.


```python
import numpy as np
import pandas as pd
```

```python
#create series
s = pd.Series([1,2,3,4,5,6,7])
print(s)
```

Descriptive Statistics

**s.describe()**: This method gives a statistical summary of the Series, including count, mean, standard deviation, minimum, maximum, and quartile values.


```python

print(s.describe())
```

# **Data Manipulation**

**s.map(func)**: Applies a function to each element in the Series.


```python
mapping = s.map(lambda x :x*2)
print(mapping)
print(s)
```

**s.apply(func)**: Similar to map, but more flexible. (Can be used Data Frames as well, where as map is only for Series)

```python
applies = s.apply(lambda x : x ** 2)
print(applies)
print(s)
```

**s.sort_values()**: Sorts the Series.

```python
s1 = pd.Series([4,7,54,23,92,41])
sorted = s1.sort_values()
print(sorted)
```

**s.drop(index)**: Drops specified labels from the Series.(not deletes)

```python
dropped = s.drop(1)
print(dropped)
print(s)
```

# **Handling Missing Data**

**s.isnull()**: Checks for missing values, returns a Series of booleans.

**s.notnull()**: Opposite of isnull().

**s.fillna(value)**: Fills missing values with a specified value.

**s.dropna()**: Drops all rows that contain missing values.like None



```python
s2 = pd.Series([2,4,6,None,np.nan,7,0])
print(s2.isnull())

print(s2.notnull())

print(s2.fillna(0))

print(s2)
print(s2.dropna())
```

# Indexing, Slicing, and Filtering

**s.iloc[ ]**: Purely integer-location based indexing. (+)-left to right,(-)-right to left

```python
print(s.iloc[2])  #1,2,3,4,5,6,7
print(s.iloc[-2])
```

**s.loc[ ]**: Label-based indexing.(only we can give positive values not negitive)

```python
print(s.loc[2])
print(s.loc[1:2])
```

**s[s > n]**: Filters and returns elements greater than n.

```python
greater_num = s[s<=5]
print(greater_num)
```

# Aggregation

**s.sum():** Sums up the values

**s.cumsum()**: cumulative Sum

**s.aggregate(func):** Aggregates using one or more operations. like s.aggregate(['sum','mean','std'])

```python
print(s.sum())

print(s.cumsum())

print(s.aggregate(['sum','std','mean']))
```

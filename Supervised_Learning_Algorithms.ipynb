# **1.Simple Linear Regression (Regression)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({

    "Area":[1000,1200,1300,1400,1500],
    "Price":[2,4,6,8,10]
})

X = df[["Area"]]

y = df["Price"]

model = LinearRegression()

model.fit(X,y)

prediction = model.predict([[1800]])

print("Predicted Salary:",prediction[0])

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)

plt.scatter(df["Area"], df["Price"])
plt.title("House Price prediction")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```


# **2.Multiple Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df = pd.DataFrame({
    "Mileage": [10, 15, 18, 20, 25],
    "Engine": [1200, 1400, 1600, 1800, 2000],
    "Age": [1, 2, 3, 4, 5],
    "Price": [8, 7, 6, 5, 4]
})

X = df[["Mileage", "Engine", "Age"]]

y = df["Price"]

model = LinearRegression()

model.fit(X, y)

prediction = model.predict([[16, 1500, 2]])

print("Predicted Car Price:", prediction[0])
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

# Visualization
plt.figure(figsize=(7,5))
plt.bar(df["Mileage"], df["Price"])

plt.title("Car Price Prediction")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```

# **3.Decision Tree(Classification & Regression)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

df = pd.DataFrame({
    "BloodSugar": [80, 90, 100, 120, 150, 180],
    "Diabetes": [0, 0, 0, 1, 1, 1]
})

X = df[["BloodSugar"]]
y = df["Diabetes"]

model = DecisionTreeClassifier()

model.fit(X, y)

prediction = model.predict([[140]])

if prediction[0] == 1:
    print("Prediction: Diabetes")
else:
    print("Prediction: No Diabetes")

plt.figure(figsize=(6,4))
plt.bar(df["BloodSugar"], df["Diabetes"])

plt.title("Blood Sugar vs Diabetes")
plt.xlabel("Blood Sugar")
plt.ylabel("Diabetes")

plt.show()
```

# **4.Logistic Regression(Classification)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

df = pd.DataFrame({
    "Income": [20, 25, 30, 40, 50, 60],
    "Subscribed": [0, 0, 0, 1, 1, 1]
})

X = df[["Income"]]
y = df["Subscribed"]

model = LogisticRegression()
model.fit(X, y)

prediction = model.predict([[45]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Customer Subscribed")
else:
    print("Customer Not Subscribed")

plt.plot(df["Income"], df["Subscribed"])
plt.title("Customer Subscription")
plt.xlabel("Income")
plt.ylabel("Subscribed")
plt.grid(True)
plt.show()
```

# **5.Naive Bayes**

```python
# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

df = pd.DataFrame({
    "Review": ["Good","Excellent","Bad","Amazing","Worst"],
    "Sentiment": ["Positive","Positive","Negative","Positive","Negative"]})

cv = CountVectorizer()
X = cv.fit_transform(df["Review"])

y = df["Sentiment"]

model = MultinomialNB()

model.fit(X, y)

test = cv.transform(["Excellent"])

prediction = model.predict(test)

print("Prediction:", prediction[0])

count = df["Sentiment"].value_counts()

plt.pie(count.values,
        labels=count.index,
        autopct="%1.1f%%")

plt.title("Sentiment Analysis")
plt.show()
```

# **6.K-Nearest Neighbors(KNN)**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

df = pd.DataFrame({
    "Weight":[3,4,5,20,25,30,35,45, ],
    "Animal":["Cat","Cat","Cat","Dog","Dog","Dog","Rabbit","Rabbit"]
})

X = df[["Weight"]]
y = df["Animal"]

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)

prediction = model.predict([[22]])

print("Prediction:", prediction[0])

# Pie Chart
count = df["Animal"].value_counts()

plt.pie(count.values,
        labels=count.index,
        autopct="%1.1f%%")

plt.title("Animal Distribution")
plt.show()
```

# **7.Random Forest**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

df = pd.DataFrame({
    "Experience": [1, 2, 3, 5, 6, 8],
    "Promotion": [0, 0, 0, 1, 1, 1]
})

X = df[["Experience"]]
y = df["Promotion"]

model = RandomForestClassifier(n_estimators=10, random_state=42)
model.fit(X, y)

prediction = model.predict([[4]])

print("Prediction:", prediction[0])

if prediction[0] == 1:
    print("Employee Promoted")
else:
    print("Employee Not Promoted")

count = df["Promotion"].value_counts()

plt.plot(df["Experience"], df["Promotion"], marker="o")

plt.title("Experience vs Promotion")
plt.xlabel("Experience (Years)")
plt.ylabel("Promotion (0 = No, 1 = Yes)")

plt.grid(True)
plt.show()
```

# **8.Support Vector Machine(SVM)**

```python
import pandas as pd
from sklearn.svm import SVC

df = pd.DataFrame({
    "Length": [8, 10, 12, 20, 22, 25],
    "Vegetable": ["Carrot", "Sweet potato", "Carrot", "Cucumber", "Cucumber", "Cucumber"]
})

X = df[["Length"]]
y = df["Vegetable"]

model = SVC(kernel="linear")
model.fit(X, y)

prediction = model.predict([[18]])

print("Prediction:", prediction[0])

count = df["Vegetable"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(count.values,
        labels=count.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Vegetable Distribution")
plt.show()
```

# **write a program on detecting whether the mail is spam or not**

```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

df = pd.DataFrame({
    "Email": [
        "Win money now",
        "Free gift",
        "Meeting today",
        "Project update",
        "Claim your prize",
        "Team discussion"
    ],
    "Spam": ["Yes", "Yes", "No", "No", "Yes", "No"]
})

cv = CountVectorizer()
X = cv.fit_transform(df["Email"])

y = df["Spam"]

model = LogisticRegression()

model.fit(X, y)

test = cv.transform(["Free prize"])
prediction = model.predict(test)

print("\nPrediction:", prediction[0])

if prediction[0] == "Yes":
    print("a Spam Mail")
else:
    print("Not a Spam Mail")

count = df["Spam"].value_counts()

plt.bar(count.index, count.values)

plt.title("Spam Email Distribution")
plt.xlabel("Email Type")
plt.ylabel("Count")

plt.show()
```
